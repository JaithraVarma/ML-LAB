```python
def cont_to_exp(x):
    if type(x) is str:
        x = x.replace('\\','')
        for key in contractions:
            value = contractions[key]
            x = x.replace(key, value)
        return x
    else:
        return x
```

```python
x = "Hello what're you doing. I can't don't this"
print(cont_to_exp(x))
```

```python
# applying the function to all review statements 
df["Review Text"] = df["Review Text"].apply(lambda x:cont_to_exp(x))
df['Review Text']
```

### Separate the columns into dependent and independent variables (or features and labels). Then you split those variables into train and test sets (80:20).

```python
list_in = df.drop(['Recommended IND','Title','Review Text','Division Name','Department Name','Class Name'],axis='columns')
```

```python
dependent = df['Recommended IND']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(list_in,dependent,test_size = 0.2)
```

### Apply the Naïve Bayes Classification Algorithm on Sentiment category to predict if item is recommended

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```

### Tabulate accuracy in terms of precision,recall and F1 score.

```python
precision = cm[1,1]/(cm[1,1] + cm[0,1])
precision
```

```python
recall = cm[1,1]/(cm[1,1] + cm[1,0])
recall
```

```python
F1 = 2*precision*recall/(precision + recall)
F1
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
